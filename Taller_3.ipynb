# Funcion plot ( plt.plot() )
La funcion plot nos permite realizar graficos de lineas de manera muy sencilla

**Sintaxis Básica**

plt.plot(x,y,[formato].**Kwargs

# TIPOS DE GRÁFICOS EN MATPLOTLIB

## FUNCIONES PRINCIPALES PARA DIFERENTES TIPOS DE VISUALIZACIONES

| Función | Tipo de Gráfico | Característica Principal | Uso | Ejemplo |
|---------|-----------------|--------------------------|-----|---------|
| **`plt.plot(x, y)`** | Líneas (Line Plot) | Muestra la evolución de datos a lo largo de una variable continua | Ideal para series temporales o tendencias | `plt.plot([1,2,3], [4,5,6])` |
| **`plt.scatter(x, y)`** | Dispersión (Scatter Plot) | Muestra la relación entre dos variables. Cada punto es un par (x, y) | Ideal para detectar correlaciones o outliers | `plt.scatter(x, y, c=colors, s=sizes)` |
| **`plt.bar(x, height)`** | Barras Verticales (Bar Plot) | Representa datos categóricos con barras rectangulares proporcionales a sus valores | Ideal para comparar cantidades entre categorías | `plt.bar(categorias, valores)` |
| **`plt.barh(y, width)`** | Barras Horizontales (Horizontal Bar Plot) | Representa datos categóricos con barras horizontales proporcionales | Ideal para etiquetas largas o muchas categorías | `plt.barh(categorias, valores)` |
| **`plt.hist(x, bins)`** | Histograma (Histogram) | Muestra la distribución de una variable numérica. Agrupa valores en "bins" (contenedores) | Ideal para entender la forma de una distribución de datos | `plt.hist(datos, bins=20)` |
| **`plt.pie(x, labels)`** | Circular (Pie Chart) | Muestra proporciones de un total. Cada segmento es proporcional a la cantidad | Ideal para mostrar porcentajes del todo (usar con precaución) | `plt.pie(sizes, labels=labels)` |
| **`plt.boxplot(x)`** | Caja y Bigotes (Box Plot) | Muestra la distribución de datos basada en cinco números: mínimo, Q1, mediana, Q3 y máximo | Ideal para comparar distribuciones y detectar outliers | `plt.boxplot([datos1, datos2])` |
| **`plt.violinplot(x)`** | Violín (Violin Plot) | Combina un box plot con una estimación de densidad de probabilidad | Ideal para mostrar la distribución completa de los datos | `plt.violinplot(dataset)` |
| **`plt.stem(x, y)`** | Tallo y Hojas (Stem Plot) | Muestra datos discretos con líneas verticales desde una línea base | Ideal para señales discretas o secuencias | `plt.stem(x, y)` |
| **`plt.step(x, y)`** | Escalones (Step Plot) | Muestra datos con cambios abruptos, como una función escalonada | Ideal para datos que cambian en puntos específicos | `plt.step(x, y, where='mid')` |
| **`plt.fill_between(x, y1, y2)`** | Área (Area Plot) | Rellena el área entre dos curvas o entre una curva y un eje | Ideal para mostrar acumulación o diferencias | `plt.fill_between(x, y1, y2, alpha=0.5)` |
| **`plt.stackplot(x, y1, y2, ...)`** | Área Apilada (Stacked Area) | Muestra múltiples series apiladas unas sobre otras | Ideal para mostrar composición a lo largo del tiempo | `plt.stackplot(x, y1, y2, y3)` |
| **`plt.contour(X, Y, Z)`** | Contorno (Contour Plot) | Muestra líneas de nivel de una función en 2D | Ideal para representar superficies 3D en 2D | `plt.contour(X, Y, Z, levels=10)` |
| **`plt.contourf(X, Y, Z)`** | Contorno Relleno (Filled Contour) | Muestra áreas de nivel de una función en 2D con colores | Ideal para mapas de calor de superficies | `plt.contourf(X, Y, Z, cmap='viridis')` |
| **`plt.imshow(data)`** | Imagen/Heatmap | Muestra datos 2D como una imagen o mapa de calor | Ideal para matrices, imágenes o datos 2D densos | `plt.imshow(matriz, cmap='hot')` |
| **`plt.pcolor(X, Y, Z)`** | Cuasi-color (Pseudo-color) | Crea un gráfico de cuadrícula con colores para valores 2D | Ideal para datos irregulares en 2D | `plt.pcolor(X, Y, Z, shading='auto')` |
| **`plt.hexbin(x, y)`** | Hexbin | Muestra la densidad de puntos en una malla hexagonal | Ideal para grandes conjuntos de datos de dispersión | `plt.hexbin(x, y, gridsize=20)` |
| **`plt.errorbar(x, y, yerr)`** | Barras de Error | Muestra datos con intervalos de incertidumbre | Ideal para datos experimentales con errores | `plt.errorbar(x, y, yerr=error)` |
| **`plt.quiver(X, Y, U, V)`** | Campo Vectorial (Quiver) | Muestra vectores como flechas en una cuadrícula | Ideal para campos de velocidad o fuerza | `plt.quiver(X, Y, U, V, scale=20)` |
| **`plt.streamplot(X, Y, U, V)`** | Líneas de Corriente (Streamplot) | Muestra líneas de flujo de un campo vectorial | Ideal para visualizar flujos fluidos | `plt.streamplot(X, Y, U, V, density=1)` |
| **`plt.polar(theta, r)`** | Coordenadas Polares | Crea gráficos en coordenadas polares | Ideal para datos angulares o direccionales | `plt.polar(angulos, radios)` |

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 12]
datos_normales = np.random.normal(0, 1, 1000)

# Ejemplos de cada tipo
fig, axes = plt.subplots(3, 3, figsize=(12, 10))

# Línea
axes[0,0].plot(x, y)
axes[0,0].set_title('Line Plot')

# Dispersión
axes[0,1].scatter(x[:20], y[:20])
axes[0,1].set_title('Scatter Plot')

# Barras
axes[0,2].bar(categorias, valores)
axes[0,2].set_title('Bar Plot')

# Histograma
axes[1,0].hist(datos_normales, bins=20)
axes[1,0].set_title('Histogram')

# Box Plot
axes[1,1].boxplot([datos_normales])
axes[1,1].set_title('Box Plot')

# Pie Chart
axes[1,2].pie(valores, labels=categorias)
axes[1,2].set_title('Pie Chart')

# Area Plot
axes[2,0].fill_between(x, y, alpha=0.5)
axes[2,0].set_title('Area Plot')

# Stem Plot
axes[2,1].stem(x[:10], y[:10])
axes[2,1].set_title('Stem Plot')

# Errorbar
axes[2,2].errorbar(x[:5], y[:5], yerr=0.1*np.ones(5))
axes[2,2].set_title('Errorbar')

plt.tight_layout()
plt.show()
```

# **Gráficos orientado a objetos (axes)**

Es un modulo orientado a la presentación y diseño para presentar multiples gráficos de una forma simplificada, utilizando el enfoque de los arreglos en función de los ejes (axes).

**Ventajas**
- Mas Control y Organización
- Ideal para uso en aplicaciónes
- Reutilización de objetos
- Explicito y as facil de leer y análizar



```python
fig, ax = plt.subplots()
# fig es el objeto Figure (el lienzo).
# ax es el objeto Axes (el sistema de coordenadas).
```

# **¿Qué es un subplot?**

Un subplot es una forma de organizar múltiples gráficos (axes) dentro de una misma figura en matplotlib. Permite crear cuadrículas de gráficos para comparar visualizaciones o mostrar diferentes aspectos de los datos.

```python
# Método tradicional usando índices
plt.figure(figsize=(12, 8))

# Subplot 1 (fila 1, col 1, gráfico 1)
plt.subplot(2, 2, 1)  # (filas, columnas, índice)
plt.plot(x, np.sin(x), 'r-')
plt.title('Seno')

# Subplot 2 (fila 1, col 2, gráfico 2)
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'b-')
plt.title('Coseno')

# Subplot 3 (fila 2, col 1, gráfico 3)
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x), 'g-')
plt.title('Tangente')
plt.ylim(-5, 5)

# Subplot 4 (fila 2, col 2, gráfico 4)
plt.subplot(2, 2, 4)
plt.plot(x, np.exp(-x), 'm-')
plt.title('Exponencial Negativa')

plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Crear una figura con 2 filas y 2 columnas de subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Datos de ejemplo
x = np.linspace(0, 10, 100)

# Acceder a cada subplot individualmente
axes[0, 0].plot(x, np.sin(x), 'r-')
axes[0, 0].set_title('Seno')

axes[0, 1].plot(x, np.cos(x), 'b-')
axes[0, 1].set_title('Coseno')

axes[1, 0].plot(x, np.tan(x), 'g-')
axes[1, 0].set_title('Tangente')
axes[1, 0].set_ylim(-5, 5)

axes[1, 1].plot(x, np.exp(x), 'm-')
axes[1, 1].set_title('Exponencial')

plt.tight_layout()
plt.show()
```

# COMPARACIÓN: Pyplot (plt.) vs Métodos de Axes (ax.)

| Función Pyplot (plt.) | Método de Axes (ax.) | Descripción |
|----------------------|---------------------|-------------|
| `plt.plot(x, y)` | `ax.plot(x, y)` | Crea un gráfico de líneas |
| `plt.title('Título')` | `ax.set_title('Título')` | Establece el título del gráfico |
| `plt.xlabel('Etiqueta')` | `ax.set_xlabel('Etiqueta')` | Etiqueta el eje X |
| `plt.ylabel('Etiqueta')` | `ax.set_ylabel('Etiqueta')` | Etiqueta el eje Y |
| `plt.legend()` | `ax.legend()` | Muestra la leyenda |
| `plt.xlim(min, max)` | `ax.set_xlim(min, max)` | Establece límites del eje X |
| `plt.ylim(min, max)` | `ax.set_ylim(min, max)` | Establece límites del eje Y |
| `plt.grid(True)` | `ax.grid(True)` | Activa la cuadrícula |
| `plt.xticks(ticks)` | `ax.set_xticks(ticks)` | Define los ticks del eje X |
| `plt.yticks(ticks)` | `ax.set_yticks(ticks)` | Define los ticks del eje Y |
| `plt.xscale('log')` | `ax.set_xscale('log')` | Cambia escala del eje X |
| `plt.yscale('log')` | `ax.set_yscale('log')` | Cambia escala del eje Y |
| `plt.text(x, y, 'texto')` | `ax.text(x, y, 'texto')` | Añade texto en coordenadas |
| `plt.annotate('texto', xy, xytext)` | `ax.annotate('texto', xy, xytext)` | Añade anotación con flecha |
| `plt.axis('equal')` | `ax.axis('equal')` | Establece aspecto igual para ejes |
| `plt.cla()` | `ax.clear()` | Limpia los ejes |
| `plt.scatter(x, y)` | `ax.scatter(x, y)` | Crea gráfico de dispersión |
| `plt.bar(x, height)` | `ax.bar(x, height)` | Crea gráfico de barras |
| `plt.hist(data)` | `ax.hist(data)` | Crea histograma |
| `plt.pie(data)` | `ax.pie(data)` | Crea gráfico circular |
| `plt.imshow(data)` | `ax.imshow(data)` | Muestra imagen/matriz |
| `plt.colorbar()` | `fig.colorbar(im, ax=ax)` | Añade barra de color |
| `plt.subplot()` | `fig, ax = plt.subplots()` | Crea figura y ejes |

```python
# Configuración para mejores gráficos
plt.style.use('seaborn-v0_8')
np.random.seed(42)

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 12]
datos_normales = np.random.normal(0, 1, 1000)
datos1 = np.random.normal(0, 1, 200)
datos2 = np.random.normal(2, 1.5, 200)

# Datos para gráficos 2D/3D
X, Y = np.meshgrid(np.linspace(-3, 3, 30), np.linspace(-3, 3, 30))
Z = np.sin(X) + np.cos(Y)

# Datos para campos vectoriales
U = -Y  # Componente x del vector
V = X   # Componente y del vector

# Datos para hexbin
x_hex = np.random.normal(0, 1, 1000)
y_hex = np.random.normal(0, 1, 1000)

# Datos polares
theta = np.linspace(0, 2*np.pi, 100)
r = 2 + np.cos(5*theta)

# Crear figura grande con todos los gráficos
fig = plt.figure(figsize=(20, 20))

# 1. Line Plot
ax1 = plt.subplot(5, 5, 1)
ax1.plot(x, y, 'b-', linewidth=2)
ax1.set_title('Line Plot')
ax1.grid(True, alpha=0.3)

# 2. Scatter Plot
ax2 = plt.subplot(5, 5, 2)
colors = np.random.rand(20)
sizes = 100 * np.random.rand(20)
ax2.scatter(x[:20], y[:20], c=colors, s=sizes, alpha=0.7, cmap='viridis')
ax2.set_title('Scatter Plot')

# 3. Bar Plot Vertical
ax3 = plt.subplot(5, 5, 3)
ax3.bar(categorias, valores, color=['red', 'blue', 'green', 'orange'], alpha=0.7)
ax3.set_title('Bar Plot (Vertical)')

# 4. Bar Plot Horizontal
ax4 = plt.subplot(5, 5, 4)
ax4.barh(categorias, valores, color=['red', 'blue', 'green', 'orange'], alpha=0.7)
ax4.set_title('Bar Plot (Horizontal)')

# 5. Histogram
ax5 = plt.subplot(5, 5, 5)
ax5.hist(datos_normales, bins=30, alpha=0.7, color='purple', edgecolor='black')
ax5.set_title('Histogram')

# 6. Pie Chart
ax6 = plt.subplot(5, 5, 6)
ax6.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90)
ax6.set_title('Pie Chart')

# 7. Box Plot
ax7 = plt.subplot(5, 5, 7)
ax7.boxplot([datos1, datos2], labels=['Grupo 1', 'Grupo 2'])
ax7.set_title('Box Plot')

# 8. Violin Plot
ax8 = plt.subplot(5, 5, 8)
violin_parts = ax8.violinplot([datos1, datos2], showmeans=True)
for pc in violin_parts['bodies']:
    pc.set_facecolor('lightblue')
    pc.set_alpha(0.7)
ax8.set_title('Violin Plot')

# 9. Stem Plot
ax9 = plt.subplot(5, 5, 9)
ax9.stem(x[:15], y[:15], linefmt='b-', markerfmt='bo', basefmt='r-')
ax9.set_title('Stem Plot')

# 10. Step Plot
ax10 = plt.subplot(5, 5, 10)
ax10.step(x[:20], y[:20], where='mid', linewidth=2)
ax10.set_title('Step Plot')

# 11. Area Plot (Fill Between)
ax11 = plt.subplot(5, 5, 11)
ax11.fill_between(x, y, alpha=0.5, color='lightblue')
ax11.plot(x, y, 'b-', linewidth=1)
ax11.set_title('Area Plot')

# 12. Stacked Area
ax12 = plt.subplot(5, 5, 12)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)
ax12.stackplot(x, y1, y2, y3, alpha=0.7)
ax12.set_title('Stacked Area')

# 13. Contour Plot
ax13 = plt.subplot(5, 5, 13)
contour = ax13.contour(X, Y, Z, levels=10)
ax13.clabel(contour, inline=True, fontsize=8)
ax13.set_title('Contour Plot')

# 14. Filled Contour
ax14 = plt.subplot(5, 5, 14)
contourf = ax14.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contourf, ax=ax14)
ax14.set_title('Filled Contour')

# 15. Heatmap (imshow)
ax15 = plt.subplot(5, 5, 15)
heatmap_data = np.random.rand(10, 10)
im = ax15.imshow(heatmap_data, cmap='hot', interpolation='nearest')
plt.colorbar(im, ax=ax15)
ax15.set_title('Heatmap (imshow)')

# 16. Pcolor
ax16 = plt.subplot(5, 5, 16)
pc = ax16.pcolor(X, Y, Z, cmap='coolwarm', shading='auto')
plt.colorbar(pc, ax=ax16)
ax16.set_title('Pcolor Plot')

# 17. Hexbin
ax17 = plt.subplot(5, 5, 17)
hb = ax17.hexbin(x_hex, y_hex, gridsize=20, cmap='Blues')
plt.colorbar(hb, ax=ax17)
ax17.set_title('Hexbin Plot')

# 18. Errorbar
ax18 = plt.subplot(5, 5, 18)
x_err = x[:8]
y_err = y[:8]
y_error = 0.1 * np.ones_like(x_err)
ax18.errorbar(x_err, y_err, yerr=y_error, fmt='o-', capsize=5, capthick=2)
ax18.set_title('Errorbar')

# 19. Quiver (Campo Vectorial)
ax19 = plt.subplot(5, 5, 19)
# Reducir densidad para mejor visualización
X_q, Y_q = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))
U_q = -Y_q
V_q = X_q
ax19.quiver(X_q, Y_q, U_q, V_q, scale=20, color='red')
ax19.set_title('Quiver Plot')

# 20. Streamplot
ax20 = plt.subplot(5, 5, 20)
ax20.streamplot(X, Y, U, V, density=1.5, color='blue', linewidth=1)
ax20.set_title('Streamplot')

# 21. Polar Plot
ax21 = plt.subplot(5, 5, 21, polar=True)
ax21.plot(theta, r, 'r-', linewidth=2)
ax21.set_title('Polar Plot', pad=20)

# 22. Errorbar con x e y errors
ax22 = plt.subplot(5, 5, 22)
x_error = 0.2 * np.ones_like(x_err)
ax22.errorbar(x_err, y_err, xerr=x_error, yerr=y_error, fmt='s-',
              capsize=4, elinewidth=2, markeredgewidth=2)
ax22.set_title('Errorbar XY')

# 23. Histogram 2D
ax23 = plt.subplot(5, 5, 23)
hist2d = ax23.hist2d(x_hex, y_hex, bins=20, cmap='plasma')
plt.colorbar(hist2d[3], ax=ax23)
ax23.set_title('2D Histogram')

# 24. Gráfico de Barras de Error
ax24 = plt.subplot(5, 5, 24)
bar_heights = [np.mean(datos1), np.mean(datos2)]
bar_errors = [np.std(datos1), np.std(datos2)]
ax24.bar(['Grupo 1', 'Grupo 2'], bar_heights, yerr=bar_errors,
         capsize=10, alpha=0.7, color=['lightblue', 'lightgreen'])
ax24.set_title('Bar Plot con Error')

# 25. Gráfico de Densidad
ax25 = plt.subplot(5, 5, 25)
from scipy.stats import gaussian_kde
kde = gaussian_kde(datos_normales)
x_kde = np.linspace(-4, 4, 100)
y_kde = kde(x_kde)
ax25.plot(x_kde, y_kde, 'r-', linewidth=2)
ax25.fill_between(x_kde, y_kde, alpha=0.3, color='red')
ax25.set_title('Density Plot')

plt.tight_layout()
plt.suptitle('GALERÍA COMPLETA DE GRÁFICOS MATPLOTLIB',
             fontsize=16, fontweight='bold', y=0.98)
plt.show()

print("¡Galería completa generada! Se muestran 25 tipos diferentes de gráficos.")
print("\nTipos incluidos:")
print("1. Line Plot       2. Scatter Plot    3. Bar Plot (V)    4. Bar Plot (H)   5. Histogram")
print("6. Pie Chart       7. Box Plot        8. Violin Plot     9. Stem Plot     10. Step Plot")
print("11. Area Plot     12. Stacked Area   13. Contour        14. Filled Contour15. Heatmap")
print("16. Pcolor        17. Hexbin         18. Errorbar       19. Quiver       20. Streamplot")
print("21. Polar         22. Errorbar XY    23. 2D Histogram   24. Bar Error    25. Density Plot")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# Configuración
np.random.seed(42)
plt.style.use('seaborn-v0_8')

print("🏛️ GENERANDO DATASET: ECONOMÍA DE 'PROSPERIA'...")

# Crear dataset económico completo
n = 120  # 10 años de datos mensuales

# 1. Serie temporal (10 años de datos mensuales)
fechas = pd.date_range(start='2014-01-01', end='2023-12-31', freq='M')

# 2. Variables macroeconómicas principales
# PIB (Producto Interno Bruto) con tendencia creciente y ciclos económicos
pib = (500 + 8 * np.arange(n) +
       50 * np.sin(2 * np.pi * np.arange(n) / 12) +  # Ciclo anual
       30 * np.sin(2 * np.pi * np.arange(n) / 60) +  # Ciclo quinquenal
       np.random.normal(0, 15, n))

# Inflación (IPC)
inflacion = (2.5 + 1.5 * np.sin(2 * np.pi * np.arange(n) / 24) +
             np.random.normal(0, 0.8, n)).clip(0.5, 12)

# Tasas de interés
tasa_interes = (3.0 + 2.0 * np.sin(2 * np.pi * np.arange(n) / 36) +
                np.random.normal(0, 0.5, n)).clip(0.1, 10)

# Desempleo
desempleo = (8.0 - 0.03 * np.arange(n) +
             2 * np.sin(2 * np.pi * np.arange(n) / 48) +
             np.random.normal(0, 0.7, n)).clip(3, 15)

# 3. Comercio exterior
exportaciones = (300 + 4 * np.arange(n) +
                 40 * np.sin(2 * np.pi * np.arange(n) / 12) +
                 np.random.normal(0, 25, n))

importaciones = (320 + 4.2 * np.arange(n) +
                 35 * np.sin(2 * np.pi * np.arange(n) / 12) +
                 np.random.normal(0, 20, n))

# 4. Sectoriales (contribución al PIB)
agricultura = (60 + 0.5 * np.arange(n) +
               15 * np.sin(2 * np.pi * np.arange(n) / 12) +
               np.random.normal(0, 8, n))

industria = (180 + 4 * np.arange(n) +
             25 * np.sin(2 * np.pi * np.arange(n) / 24) +
             np.random.normal(0, 12, n))

servicios = (260 + 3.5 * np.arange(n) +
             30 * np.sin(2 * np.pi * np.arange(n) / 18) +
             np.random.normal(0, 15, n))

# 5. Finanzas públicas
deuda_publica = (40 + 0.8 * np.arange(n) +
                 0.5 * np.sin(2 * np.pi * np.arange(n) / 60) +
                 np.random.normal(0, 2, n))

deficit_fiscal = (3.0 + 1.0 * np.sin(2 * np.pi * np.arange(n) / 24) +
                  np.random.normal(0, 0.8, n))

# 6. Mercado laboral por sectores
empleo_agricultura = np.random.normal(15, 2, n)
empleo_industria = np.random.normal(25, 3, n)
empleo_servicios = np.random.normal(50, 4, n)

# 7. Indicadores sociales
salario_promedio = (1500 + 30 * np.arange(n) +
                    100 * np.sin(2 * np.pi * np.arange(n) / 24) +
                    np.random.normal(0, 80, n))

pobreza = (25 - 0.2 * np.arange(n) +
           3 * np.sin(2 * np.pi * np.arange(n) / 36) +
           np.random.normal(0, 1.5, n)).clip(8, 35)

# 8. Datos por región (para comparativas)
regiones = ['Capital', 'Norte', 'Sur', 'Este', 'Oeste']
crecimiento_regional = {
    'Capital': np.random.normal(3.2, 0.8, n),
    'Norte': np.random.normal(2.8, 1.2, n),
    'Sur': np.random.normal(2.5, 1.0, n),
    'Este': np.random.normal(3.0, 0.9, n),
    'Oeste': np.random.normal(2.6, 1.1, n)
}

# 9. Datos para gráficos de torta (composición)
composicion_pib_2023 = {
    'Agricultura': 12.5,
    'Industria': 28.3,
    'Servicios': 45.2,
    'Construcción': 8.7,
    'Minería': 5.3
}

composicion_exportaciones = {
    'Manufacturas': 45,
    'Agroindustria': 25,
    'Minerales': 15,
    'Servicios': 12,
    'Otros': 3
}

# 10. Datos para heatmap (correlación entre variables)
variables_macro = ['PIB', 'Inflación', 'Desempleo', 'Exportaciones', 'Salario']

# 11. Datos para box plots (distribución por año)
años = list(range(2014, 2024))
pib_anual = [np.random.normal(500 + i*8, 20) for i in range(10)]

# Crear DataFrame principal
df_economia = pd.DataFrame({
    'Fecha': fechas,
    'Año': fechas.year,
    'Mes': fechas.month,
    'Trimestre': fechas.quarter,
    'PIB_Millones': pib,
    'Inflacion_Anual': inflacion,
    'Tasa_Interes': tasa_interes,
    'Desempleo_Porcentaje': desempleo,
    'Exportaciones_Millones': exportaciones,
    'Importaciones_Millones': importaciones,
    'Agricultura_Millones': agricultura,
    'Industria_Millones': industria,
    'Servicios_Millones': servicios,
    'Deuda_Publica_Porcentaje_PIB': deuda_publica,
    'Deficit_Fiscal_Porcentaje': deficit_fiscal,
    'Salario_Promedio': salario_promedio,
    'Pobreza_Porcentaje': pobreza,
    'Empleo_Agricultura': empleo_agricultura,
    'Empleo_Industria': empleo_industria,
    'Empleo_Servicios': empleo_servicios
})

# Añadir datos regionales
for region, crecimiento in crecimiento_regional.items():
    df_economia[f'Crecimiento_{region}'] = crecimiento

print("✅ DATASET ECONÓMICO CREADO EXITOSAMENTE")
print(f"📊 Dimensiones: {df_economia.shape[0]} registros × {df_economia.shape[1]} variables")
print("\n🏛️ VARIABLES DISPONIBLES:")
for i, col in enumerate(df_economia.columns, 1):
    print(f"  {i:2d}. {col}")

# Mostrar primeras filas
print("\n👀 PRIMERAS 5 FILAS:")
display(df_economia.head())

# Estadísticas descriptivas
print("\n📈 ESTADÍSTICAS DESCRIPTIVAS:")
display(df_economia.describe())
```
